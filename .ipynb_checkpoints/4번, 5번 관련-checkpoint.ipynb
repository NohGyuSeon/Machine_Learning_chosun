```python
import tensorflow as tf 
from tensorflow.keras.layers import Dense, Flatten, Conv2D, Conv1D
from tensorflow.keras import Model
```

```python
'''
tf.keras.layers.Conv2D(
    filters,
    kernel_size,
    strides=(1, 1),
    padding='valid',
    data_format=None,
    dilation_rate=(1, 1),
    groups=1,
    activation=None,
    use_bias=True,
    kernel_initializer='glorot_uniform',
    bias_initializer='zeros',
    kernel_regularizer=None,
    bias_regularizer=None,
    activity_regularizer=None,
    kernel_constraint=None,
    bias_constraint=None,
    **kwargs
)
'''

data = tf.random.uniform(shape=(2, 16, 16, 3))
data = tf.cast(data, 'float32')
```

```python
class conv2d_test(Model):
    def __init__(self):
        super(conv2d_test, self).__init__()
        # filter * Kernel_size * input_channel + filters(bias)
        # 4x5x5x3 + 4
        self.conv1 = Conv2D(filters=4, kernel_size=(5,5), strides=(1,1), padding='same', activation='relu')
        # 5x3x3x4 + 5
        self.conv2 = Conv2D(filters=5, kernel_size=3, strides=2, padding='same', activation='relu')
        # 6x5x3x5 + 6
        self.conv3 = Conv2D(filters=6, kernel_size=(5,3), strides=1, padding='valid', activation='relu')    
    def call(self, x):
        x1 = self.conv1(x)
        x2 = self.conv2(x1)
        x3 = self.conv3(x2)
        return x, x1, x2, x3
       
cmodel = conv2d_test()
x, x1, x2, x3 = cmodel(data) #call 함수 호출. forwardpass 호출.
print(x.shape)
print(x1.shape)
print(x2.shape)
print(x3.shape)
```

```python
model2 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(16,16,3)), 
    tf.keras.layers.Reshape((16,16,3)), 
    Conv2D(filters=4, kernel_size=(5,5), strides=(1,1), padding='same', activation='relu'),
    Conv2D(filters=5, kernel_size=3, strides=2, padding='same', activation='relu'),
    Conv2D(filters=6, kernel_size=(5,3), strides=1, padding='valid', activation='relu')     
]) 
model2.summary()
```

```python
model3 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(24,24,3)), 
    tf.keras.layers.Reshape((24,24,3)), 
    Conv2D(filters=4, kernel_size=(5,5), strides=(1,1), padding='same', activation='relu'),
    Conv2D(filters=5, kernel_size=3, strides=2, padding='same', activation='relu'),
    Conv2D(filters=6, kernel_size=(5,3), strides=1, padding='valid', activation='relu')     
]) 
model3.summary()

yp = model3(x_train1)
print(yp)
loss_fn = tf.losses.CategoricalCrossentropy(from_logits = False)

myoptimizer = tf.optimizers.SGD(learning_rate = 0.01)
model3.compile(optimizer=myoptimizer, loss=loss_fn, metrics=['accuracy'])

model3.summary()
```
