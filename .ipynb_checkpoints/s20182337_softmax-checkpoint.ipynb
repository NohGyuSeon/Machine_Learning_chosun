```python
import tensorflow as tf
import numpy as np 
```

```python
x_train = tf.random.uniform(shape=(6, 4))
y_train = tf.random.uniform(shape=(6, 1), minval=0, maxval=3, dtype=tf.dtypes.int32)
# 6x1, int32 값 0, 1, 2
print(y_train)
x_train_one_hot = tf.one_hot(y_train, depth=3)
y_train_one_hot = tf.reshape(x_train_one_hot, shape=(-1,3))# -1은 알아서 맞추어 주세요
print(y_train_one_hot) 

x_test = tf.random.uniform(shape=(64, 4))
y_test = tf.random.uniform(shape=(64, 1), minval=0, maxval=3, dtype=tf.dtypes.int32)
y_test_one_hot = tf.one_hot(y_test, depth=3)
y_test_one_hot = tf.reshape(y_test_one_hot, shape=(-1,3))
```

```python
model = tf.keras.models.Sequential([
    tf.keras.Input(shape=(4,)),
    tf.keras.layers.Dense(units=16, activation='relu'),
    tf.keras.layers.Dense(units=3, activation=None),
    tf.keras.layers.Softmax()
])
model.summary()
yp = model(x_train)
print(yp)
```

```python
loss_fn = tf.keras.losses.CategoricalCrossentropy(from_logits = False)
optimizer = tf.keras.optimizers.SGD(learning_rate = 0.01)
model.compile(optimizer=optimizer, loss=loss_fn, metric=['accuarcy'])
model.fit(x_train, y_train_one_hot, epochs=100, batch_size=6) 
```

```python
model.evaluate(x_train, y_train_one_hot, verbose=2)
model.evaluate(x_test, y_test_one_hot, verbose=2)
```
